```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

labels = ["NAML", "DKN", "NPA", "LSTUR", "NRMS"]

group_auc = [0.655, 0.592, 0.6001, 0.6428, 0.6095]
mean_mrr = [0.3033, 0.2008, 0.2716, 0.2985, 0.2687]
ndcg_5 = [0.3333, 0.2035, 0.2941, 0.3314, 0.2887]
ndcg_10 = [0.3994, 0.2727, 0.3599, 0.3929, 0.3606]
    
x = np.arange(len(labels))  
width = 0.1  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width*2, group_auc, width, label='group_auc')
rects2 = ax.bar(x - width+0.01, mean_mrr, width, label='mean_mrr')
rects3 = ax.bar(x + 0.02, ndcg_5, width, label='ndcg_5')
rects4 = ax.bar(x + width+ 0.03, ndcg_10, width, label='ndcg_10')


ax.set_ylabel('metric performance', fontsize=16)
ax.set_xlabel('model', fontsize=16)
ax.set_title('result analysis')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(loc=2, bbox_to_anchor=(1.05,1.0),borderaxespad = 0.)


fig.tight_layout()

plt.show()
```
